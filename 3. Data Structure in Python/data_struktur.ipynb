# Linked List

Linked List adalah struktur data linear yang terdiri dari node-node yang saling terhubung lewat pointer (atau referensi). Tidak seperti array, elemen-elemen di linked list tidak disimpan secara berurutan di memori, tetapi saling menunjuk satu sama lain.

üîÅ Cara Kerja
Misalnya kita punya singly linked list (arah 1 arah):

1. Kamu mulai dari head (simpul pertama)

2. Untuk mencapai node ke-3, kamu harus lewati node ke-1 dan ke-2 dulu

3. Waktu akses: O(n) (lebih lambat dibanding array untuk akses langsung)

```python
liked_list = ["one", "two", "three"]
print(liked_list)
print()

# menambahkan elemen ke akhir list
liked_list.append("four")
liked_list.insert(0, "nol")
print(liked_list)
print()

# menghapus elemen list
liked_list.remove("nol")
print(liked_list)
print()
```

```python
liked_list = ["first", "second", "third"]
print(liked_list)
print()

# menambah element
liked_list.append("fourth")
liked_list.append("fifth")
liked_list.insert(3, "sixth")
print(liked_list)
liked_list.insert(0, "zero")
print(liked_list)
print()

# menghapus element
liked_list.remove("second")
print(liked_list)
print()
```

# Stack

Stack adalah struktur data linear yang mengikuti prinsip LIFO (Last In, First Out) ‚Äî artinya elemen terakhir yang dimasukkan adalah elemen pertama yang dikeluarkan.

üì¶ Contoh Kasus Nyata:
Bayangkan tumpukan piring:

1. Kamu menaruh piring ke atas (push)

2. Saat mengambil, kamu ambil piring paling atas dulu (pop)

| Operasi    | Deskripsi                              | Kompleksitas |
| ---------- | -------------------------------------- | ------------ |
| `push`     | Tambah elemen ke atas tumpukan         | O(1)         |
| `pop`      | Ambil elemen paling atas               | O(1)         |
| `peek`     | Lihat elemen paling atas (tanpa hapus) | O(1)         |
| `is_empty` | Cek apakah stack kosong                | O(1)         |


```python
stack = ["first", "second", "third"]
print(stack)
print()

# pushing element
stack.append("fourth")
stack.append("fifth")
print(stack)
print()

# ngambil element yang paling terakhir
print(stack[-1])
print()

# Fungsi pop() pada stack digunakan untuk menghapus dan mengambil elemen paling atas dari tumpukan (sesuai prinsip LIFO: Last In, First Out).
stack.pop()
print(stack)
```

```python
print()
stack = ["first", "second", "third", "fourth"]
print(stack)

stack.append("fifth")
print(stack)

print(stack[-1])

stack.pop()
print(stack)
print()

stack.remove(stack[2])
print(stack)
print()
```

# Graph

Struktur graph (graf) itu kayak jaringan titik-titik yang terhubung ‚Äî mirip peta jalan, sosial media, atau koneksi antar komputer.

üß† Penjelasan Sederhana:
1. Node (disebut juga vertex): titik-titiknya, misalnya kota, orang, atau komputer

2. Edge (sisi): garis penghubung antar node, misalnya jalan antar kota, pertemanan, atau kabel

```python
class adjnode:
    def __init__(self, val):
        self.val = val
        self.next = None


class graph:
    def __init__(self, vertices):
        self.v = vertices
        self.ele = [None]*self.v

    def edge(self, src, dest): #src (source), dest (destination)
        node = adjnode(dest)
        node.next = self.ele[src]
        self.ele[src] = node

        node = adjnode(src)
        node.next = self.ele[dest]
        self.ele[dest] = node

    def __repr__(self):
        for i in range(self.v):
            print("Adjacency list of vertex {}\n head".format(i), end="")
            temp = self.ele[i]
            while temp:
                print(" -> {}".format(temp.val), end="")
                temp = temp.next

g = graph(4) #mendefinisikan jumlah vertex yg ada
#menghubungkan tia vertex tsb
g.edge(0, 2) #node 0 akan bertetangga dengan 2 dst.
g.edge(1, 3)
g.edge(3, 2)
g.edge(0, 3)
g.__repr__()
```

# Hashmap

üß† Apa itu HashMap?
HashMap (kadang disebut dictionary di Python, object di JavaScript, atau map di Java/C++) adalah struktur data pasangan kunci-nilai (key-value).

```python
 #mendefiniskan fungsi untuk memanggil TIAP key dan valuenya
def printdict(d):
    for key in d:
        print(key, "->", d[key])
#apabila suatu dict di-print/panggil langsung (naked) dalam fungsi, yg kepanggil adalah key-nya

hashmaps = {0: 'first', 1: 'second', 2: 'third'}
printdict(hashmaps)
print()



hashmaps[4] = 'fifth'
printdict(hashmaps)
print()

hashmaps.popitem() #menghilangkan indeks dimulai dr indeks terakhir
printdict(hashmaps)
```
