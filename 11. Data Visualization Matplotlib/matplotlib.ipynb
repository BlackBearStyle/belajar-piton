## ‚úÖ **Menetapkan Batas dan Label Centang (Ticks) di `matplotlib`**

Dalam `matplotlib`, **batas** dan **centang (ticks)** digunakan untuk **mengatur skala sumbu X dan Y**, serta **memberi label** pada titik-titik tertentu agar grafik lebih jelas dibaca.

---

## üìå 1. **Batas (Limits)**

Digunakan untuk **mengatur area tampilan** sumbu.

### Contoh:

```python
plt.xlim(0, 10)  # batas sumbu X dari 0 sampai 10
plt.ylim(0, 100) # batas sumbu Y dari 0 sampai 100
```

---

## üìå 2. **Centang (Ticks)**

Centang/tick adalah **garis kecil dan label angka** di sepanjang sumbu.

### Contoh:

```python
plt.xticks([0, 2, 4, 6, 8, 10])  # atur posisi centang sumbu X
plt.yticks([0, 25, 50, 75, 100]) # atur posisi centang sumbu Y
```

### Bisa juga ditambah label:

```python
plt.xticks([1, 2, 3], ['Satu', 'Dua', 'Tiga'])
```

---

## üé® Contoh Lengkap:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y)

# Menetapkan batas
plt.xlim(0, 6)
plt.ylim(0, 40)

# Menetapkan centang
plt.xticks([1, 2, 3, 4, 5], ['Satu', 'Dua', 'Tiga', 'Empat', 'Lima'])
plt.yticks([0, 10, 20, 30, 40])

plt.title("Contoh Grafik dengan Batas & Ticks")
plt.show()
```

# Perbedaan `ticks` dan `lim`

### ‚úÖ **Perbedaan antara `ticks` dan `lim` di Matplotlib:**

#### 1. **`lim` ‚Üí `xlim()` / `ylim()`**

* Digunakan untuk **mengatur batas (limit)** sumbu X atau Y.
* Contoh:

  ```python
  plt.xlim(0, 10)  # Atur batas sumbu X dari 0 sampai 10
  plt.ylim(-1, 1)  # Atur batas sumbu Y dari -1 sampai 1
  ```
* Fungsinya: menentukan rentang data yang ditampilkan di grafik.

#### 2. **`ticks` ‚Üí `xticks()` / `yticks()`**

* Digunakan untuk **mengatur posisi dan label dari ticks (garis kecil/angka)** di sepanjang sumbu.
* Contoh:

  ```python
  plt.xticks([0, 2, 4, 6, 8, 10])  # Menampilkan ticks hanya di posisi tertentu
  plt.yticks([-1, 0, 1])           # Menentukan posisi ticks di sumbu Y
  ```
* Fungsinya: mengatur di mana angka atau label muncul di sumbu.

---

### üîç Singkatnya:

| Fungsi                  | Tujuan                          | Contoh                    |
| ----------------------- | ------------------------------- | ------------------------- |
| `xlim()` / `ylim()`     | Mengatur batas tampilan sumbu   | `plt.xlim(0, 100)`        |
| `xticks()` / `yticks()` | Mengatur posisi dan label ticks | `plt.xticks([0, 25, 50])` |

---

## üß± Apa Itu `Figure` di Matplotlib?

### üîπ `Figure` adalah:

Objek **utama** dalam Matplotlib yang merepresentasikan **keseluruhan area gambar (canvas)**, termasuk:

* Sumbu-sumbu grafik (axes)
* Judul umum (suptitle)
* Teks global
* Legenda global
* Elemen visual lainnya

---

## ‚öôÔ∏è Struktur Dasar Objek Visual Matplotlib

```
Figure
‚îî‚îÄ‚îÄ Axes (bisa lebih dari satu)
     ‚îú‚îÄ‚îÄ XAxis & YAxis
     ‚îú‚îÄ‚îÄ Title
     ‚îú‚îÄ‚îÄ Grid, Ticks, Line, etc.
```

### Analogi:

> **Figure = lembar kertas besar**
> **Axes = kotak grafik di atas kertas**

---

## üìå Fungsi dan Kegunaan `Figure`

| Fungsi `Figure`                        | Penjelasan                                |
| -------------------------------------- | ----------------------------------------- |
| Menjadi **container utama**            | Semua elemen grafik ada di dalamnya       |
| Bisa menampung **beberapa `Axes`**     | Bisa buat banyak grafik dalam satu figure |
| Kontrol **ukuran dan resolusi** gambar | Misal: `figsize`, `dpi`, dll              |
| Simpan ke file                         | Menggunakan `fig.savefig('nama.png')`     |

---

## üõ†Ô∏è Cara Membuat Figure

### 1. Dengan `plt.figure()` + `add_subplot()`

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6, 4))  # buat figure dengan ukuran 6x4 inch
ax = fig.add_subplot(1, 1, 1)     # tambah axes 1 baris, 1 kolom, posisi ke-1
ax.plot([1, 2, 3], [4, 5, 6])
fig.suptitle("Contoh Figure dan Axes")  # judul global
plt.show()
```

### 2. Dengan `plt.subplots()` (cara modern dan paling umum)

```python
fig, ax = plt.subplots(figsize=(6, 4))  # fig: figure, ax: satu axes
ax.plot([1, 2, 3], [4, 1, 2])
fig.suptitle("Grafik dengan plt.subplots()")
```

---

## üß™ Beberapa Method Penting dari `Figure`

| Method                         | Fungsi                                            |
| ------------------------------ | ------------------------------------------------- |
| `fig.add_subplot(...)`         | Tambahkan axes ke figure                          |
| `fig.add_axes([x0, y0, w, h])` | Tambah axes manual (dalam unit relatif ke figure) |
| `fig.suptitle("...")`          | Tambahkan judul global untuk figure               |
| `fig.savefig("file.png")`      | Simpan figure ke file gambar                      |
| `fig.set_size_inches(w, h)`    | Atur ukuran figure                                |

---

## üé® Contoh: Multi-Plot dengan Satu `Figure`

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(8, 6))  # 2x2 axes dalam satu figure

axs[0, 0].plot([1, 2], [3, 4])
axs[0, 1].bar([1, 2], [5, 2])
axs[1, 0].scatter([1, 2], [4, 1])
axs[1, 1].hist([1, 1, 2, 3, 3, 3, 4])

fig.suptitle("Empat Grafik dalam Satu Figure")
plt.tight_layout()
plt.show()
```

---

## üí° Tips

* Gunakan `plt.subplots()` untuk lebih rapi dan fleksibel, terutama saat membuat banyak grafik.
* Atur ukuran figure dengan `figsize=(width, height)`.
* Simpan hasil visualisasi dengan `fig.savefig("output.png")`.

## Axes Class

class Axes adalah unit paling dasar dan fleksibel untuk membuat sub-plot. Gambar tertentu mungkin berisi banyak sumbu, tetapi sumbu tertentu hanya dapat ada dalam satu gambar. Fungsi axes() membuat objek axes.

syntax:

axes([left, bottom, width, height])

````python
# Python program to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

fig = plt.figure(figsize = (5, 4))

# Adding the axes to the figure
ax = fig.add_axes([1, 1, 1, 1])

# plotting 1st dataset to the figure
ax1 = ax.plot(x, y)

# plotting 2nd dataset to the figure
ax2 = ax.plot(y, x)

# Setting Title
ax.set_title("Grafik Linear")

# Setting Label
ax.set_xlabel("Sumbu X")
ax.set_ylabel("Sumbu Y")

# Adding Legend
ax.legend(labels = ('garis 1', 'garis 2'))

plt.show()
````

## üéØ Apa Itu **Multi Plot**?

**Multi plot** itu artinya kamu membuat **banyak grafik (plot) dalam satu figure** atau satu gambar. Jadi, kamu bisa lihat beberapa grafik sekaligus, misalnya untuk perbandingan data atau visualisasi yang lebih kompleks.

---

## Kenapa pakai multi plot?

* Supaya **bisa bandingin data** di grafik berbeda secara berdampingan.
* Biar hasil visualisasi jadi **lebih rapi dan efisien**.
* Memudahkan analisis kalau ada banyak variabel.

---

## üîß Cara buat multi plot di Matplotlib

### 1. Pakai `plt.subplot()`

Contohnya bikin 2 plot dalam 1 figure:

```python

plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Plot 1")

plt.plot([1, 2, 3], [6, 5, 4])

```

---

```python
fig, axs = plt.subplots(2, 2)  # buat 4 plot (2 baris x 2 kolom)

axs[1, 1].set_title("Plot 4")

```

---

## üîë Intinya:

* **Multi plot** = banyak grafik dalam satu gambar.
* Memudahkan membandingkan data.
* Bisa diatur tata letaknya (baris, kolom, ukuran).



## üß© Apa itu `subplot2grid`?

`subplot2grid` adalah fungsi untuk **menempatkan subplot (axes) di posisi tertentu dalam grid yang fleksibel**. Dengan ini, kamu bisa mengatur layout subplot dengan ukuran dan posisi yang lebih custom dibanding `plt.subplot()` biasa.

---

## üîç Kenapa pakai `subplot2grid`?

* Kalau kamu ingin subplot yang **tidak harus sama ukuran**, misal ada subplot yang lebih besar, yang lain lebih kecil.
* Bisa membuat **layout grid kompleks** dengan kolom dan baris yang berbeda-beda untuk tiap subplot.
* Lebih fleksibel daripada `plt.subplot()` yang hanya bisa satu kotak per subplot.

---

## ‚öôÔ∏è Sintaks:

```python
plt.subplot2grid(shape, loc, rowspan=1, colspan=1, fig=None)
```

* `shape`: tuple `(rows, cols)` ‚Äî ukuran grid secara keseluruhan.
* `loc`: tuple `(row, col)` ‚Äî posisi awal subplot dalam grid.
* `rowspan`: berapa baris yang akan ditempati subplot ini (default 1).
* `colspan`: berapa kolom yang akan ditempati subplot ini (default 1).
* `fig`: opsional, jika ingin memasukkan subplot ke `Figure` tertentu.

---

## üìå Contoh sederhana:

```python
import matplotlib.pyplot as plt

fig = plt.figure()

# Grid 3x3, subplot di baris 0 kolom 0 (posisi kiri atas)
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title("Plot besar di atas")

# Subplot di baris 1 kolom 0, mengambil 1 kolom dan 2 baris
ax2 = plt.subplot2grid((3, 3), (1, 0), rowspan=2)
ax2.bar([1, 2], [3, 5])
ax2.set_title("Plot kiri bawah")

# Subplot di baris 1 kolom 1
ax3 = plt.subplot2grid((3, 3), (1, 1))
```

## üîë Kelebihan `subplot2grid`:

* Layout **lebih fleksibel dan dinamis**.
* Bisa buat subplot yang **besar dan kecil dalam satu figure**.
* Posisi subplot lebih presisi daripada `plt.subplot()`.

---


````python
import matplotlib.pyplot as plt

fig = plt.figure()

# Grid 3x3, subplot di baris 0 kolom 0 (posisi kiri atas)
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title("Plot besar di atas")

# Subplot di baris 1 kolom 0, mengambil 1 kolom dan 2 baris
ax2 = plt.subplot2grid((3, 3), (1, 0), rowspan=2)
ax2.bar([1, 2], [3, 5])
ax2.set_title("Plot kiri bawah")

# Subplot di baris 1 kolom 1
ax3 = plt.subplot2grid((3, 3), (1, 1))
ax3.scatter([1, 2, 3], [2, 4, 6])
ax3.set_title("Plot kanan tengah")

# Subplot di baris 1 kolom 2
ax4 = plt.subplot2grid((3, 3), (1, 2))
ax4.hist([1, 2, 2, 3, 3, 3])
ax4.set_title("Plot kanan bawah")

plt.tight_layout()
plt.show()
````
