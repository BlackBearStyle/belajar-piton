# MatplotLibb 

`matplotlib` adalah **library visualisasi data** di Python. Fungsinya adalah untuk **membuat grafik atau plot**, seperti grafik garis, batang, pie chart, scatter plot, histogram, dan banyak lagi.

---

## üß† Intinya:

> `matplotlib` = "Alat gambar" untuk data di Python, mirip seperti Excel chart atau grafik di Google Sheets.

---

## üé® Contoh Plot Sederhana

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Contoh Grafik")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

### üîç Output:

Grafik garis naik berdasarkan data `x` dan `y`.

---

## üì¶ Komponen Utama `matplotlib`

| Komponen        | Kegunaan                            |
| --------------- | ----------------------------------- |
| `pyplot`        | Submodul utama (seperti API gambar) |
| `plt.plot()`    | Buat garis                          |
| `plt.bar()`     | Grafik batang                       |
| `plt.scatter()` | Grafik titik                        |
| `plt.hist()`    | Histogram                           |
| `plt.pie()`     | Pie chart                           |
| `plt.show()`    | Tampilkan grafik                    |

---

## ‚úÖ **Menetapkan Batas dan Label Centang (Ticks) di `matplotlib`**

Dalam `matplotlib`, **batas** dan **centang (ticks)** digunakan untuk **mengatur skala sumbu X dan Y**, serta **memberi label** pada titik-titik tertentu agar grafik lebih jelas dibaca.

---

## üìå 1. **Batas (Limits)**

Digunakan untuk **mengatur area tampilan** sumbu.

### Contoh:

```python
plt.xlim(0, 10)  # batas sumbu X dari 0 sampai 10
plt.ylim(0, 100) # batas sumbu Y dari 0 sampai 100
```

---

## üìå 2. **Centang (Ticks)**

Centang/tick adalah **garis kecil dan label angka** di sepanjang sumbu.

### Contoh:

```python
plt.xticks([0, 2, 4, 6, 8, 10])  # atur posisi centang sumbu X
plt.yticks([0, 25, 50, 75, 100]) # atur posisi centang sumbu Y
```

### Bisa juga ditambah label:

```python
plt.xticks([1, 2, 3], ['Satu', 'Dua', 'Tiga'])
```

---

## üé® Contoh Lengkap:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y)

# Menetapkan batas
plt.xlim(0, 6)
plt.ylim(0, 40)

# Menetapkan centang
plt.xticks([1, 2, 3, 4, 5], ['Satu', 'Dua', 'Tiga', 'Empat', 'Lima'])
plt.yticks([0, 10, 20, 30, 40])

plt.title("Contoh Grafik dengan Batas & Ticks")
plt.show()
```

---

## üß† Kesimpulan:

| Fungsi     | Tujuan                      |
| ---------- | --------------------------- |
| `xlim()`   | Atur batas sumbu X          |
| `ylim()`   | Atur batas sumbu Y          |
| `xticks()` | Atur posisi & label sumbu X |
| `yticks()` | Atur posisi & label sumbu Y |

# Perbedaan `ticks` dan `lim`

### ‚úÖ **Perbedaan antara `ticks` dan `lim` di Matplotlib:**

#### 1. **`lim` ‚Üí `xlim()` / `ylim()`**

* Digunakan untuk **mengatur batas (limit)** sumbu X atau Y.
* Contoh:

  ```python
  plt.xlim(0, 10)  # Atur batas sumbu X dari 0 sampai 10
  plt.ylim(-1, 1)  # Atur batas sumbu Y dari -1 sampai 1
  ```
* Fungsinya: menentukan rentang data yang ditampilkan di grafik.

#### 2. **`ticks` ‚Üí `xticks()` / `yticks()`**

* Digunakan untuk **mengatur posisi dan label dari ticks (garis kecil/angka)** di sepanjang sumbu.
* Contoh:

  ```python
  plt.xticks([0, 2, 4, 6, 8, 10])  # Menampilkan ticks hanya di posisi tertentu
  plt.yticks([-1, 0, 1])           # Menentukan posisi ticks di sumbu Y
  ```
* Fungsinya: mengatur di mana angka atau label muncul di sumbu.

---

### üîç Singkatnya:

| Fungsi                  | Tujuan                          | Contoh                    |
| ----------------------- | ------------------------------- | ------------------------- |
| `xlim()` / `ylim()`     | Mengatur batas tampilan sumbu   | `plt.xlim(0, 100)`        |
| `xticks()` / `yticks()` | Mengatur posisi dan label ticks | `plt.xticks([0, 25, 50])` |

---

# Legend

### ‚úÖ Fungsi `legend()`

Menampilkan **keterangan** untuk setiap elemen plot yang diberi **label**.

---

### üîß Cara pakainya:

#### 1. **Dengan `label` saat plotting:**

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6], label='Data A')  # kasih label
plt.plot([1, 2, 3], [6, 5, 4], label='Data B')

plt.legend()  # tampilkan legend
plt.show()
```

#### 2. **Hasilnya:**

Di pojok plot, akan muncul kotak legend dengan tulisan:

```
‚ñ† Data A
‚ñ† Data B
```

---

### üîß Opsi tambahan:

```python
plt.legend(loc='upper left')       # posisi legend
plt.legend(title='Keterangan')     # kasih judul pada legend
```

---

### üìå Catatan:

Legend hanya muncul kalau kamu:

* Pakai `label='...'` saat plotting.
* Atau manual lewat `plt.legend(['label1', 'label2'])` (kurang disarankan karena bisa bikin gak sinkron).

---

Kalau kamu kasih kode plot-nya, bisa bantu atur legend-nya juga biar rapi.


````python
import matplotlib.pyplot as plt


# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Grafik Linear", fontsize=25, color="blue")

# Adding label on the y-axis
plt.ylabel('Sumbu Y')

# Adding label on the x-axis
plt.xlabel('Sumbu X')

# Setting the limit of y-axis
plt.ylim(0, 80)

# setting the labels of x-axis
plt.xticks(x, labels=["one", "two", "three", "four"])

# Adding legends
plt.legend(["GFG"])

plt.show()
````
