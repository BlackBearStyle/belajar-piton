# MatplotLibb 

`matplotlib` adalah **library visualisasi data** di Python. Fungsinya adalah untuk **membuat grafik atau plot**, seperti grafik garis, batang, pie chart, scatter plot, histogram, dan banyak lagi.

---

## 🧠 Intinya:

> `matplotlib` = "Alat gambar" untuk data di Python, mirip seperti Excel chart atau grafik di Google Sheets.

---

## 🎨 Contoh Plot Sederhana

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Contoh Grafik")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

### 🔍 Output:

Grafik garis naik berdasarkan data `x` dan `y`.

---

## 📦 Komponen Utama `matplotlib`

| Komponen        | Kegunaan                            |
| --------------- | ----------------------------------- |
| `pyplot`        | Submodul utama (seperti API gambar) |
| `plt.plot()`    | Buat garis                          |
| `plt.bar()`     | Grafik batang                       |
| `plt.scatter()` | Grafik titik                        |
| `plt.hist()`    | Histogram                           |
| `plt.pie()`     | Pie chart                           |
| `plt.show()`    | Tampilkan grafik                    |

---

## 🔢 Contoh Lain: Bar Chart

```python
labels = ['A', 'B', 'C']
values = [10, 15, 7]

plt.bar(labels, values)
plt.title("Nilai Kategori")
plt.show()
```
